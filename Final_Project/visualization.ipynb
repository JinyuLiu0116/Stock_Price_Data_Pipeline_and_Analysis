```python
import datetime
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import mplfinance as mpl
```

```python
df=yf.download('CRWD')
df.shape
```

```python
df.count()
```

```python
df.tail(15)
```

```python
df=df.round({'Open':4,'Close':4,'High':4,'Low':4,'Adj Close':4})
df.tail(15)
```

```python
plt.figure(figsize=(12,6))
plt.title('Price of CRWD Over The Years', fontsize=18)
plt.plot(df['Close'], label='Close price')

plt.xlabel('Date', fontsize=13)
plt.ylabel('Price In USD ($)', fontsize=13)
plt.legend()
plt.show()
```

```python
df1=df['2024-01-01':'2024-07-25']
plt.figure(figsize=(12,6))
plt.title('Price of CRWD In Year 2024', fontsize=18)
plt.fill_between(df1.index,df1['Close'],color='skyblue',alpha=0.5)
plt.plot(df1['Close'], label='Close price', marker='.')

plt.xlabel('Date', fontsize=13)
plt.ylabel('Price In USD ($)', fontsize=13)
plt.legend()
plt.show()
```

```python
fig, axs= plt.subplots(2,1, gridspec_kw={"height_ratios": [3,1]}, figsize=(13,7))
mpl.plot(df1, type="candle", ax=axs[0], style="yahoo")
axs[0].set_ylabel("Price In USD")
axs[1].bar(df1.index,df1['Volume'], color='blue')
axs[1].set_ylabel("Amount")
plt.show()
```

```python
df2=df.tail(15)
fig, axs= plt.subplots(2,1, gridspec_kw={"height_ratios": [3,1]}, figsize=(13,7))

mpl.plot(df2, type="candle", ax=axs[0], style="yahoo")
axs[1].plot(df2['Volume'], color='b')
axs[1].set_ylabel("Amount")
plt.show()
```

```python
fig, axs=plt.subplots(2,2, figsize=(13,7))

axs[0,0].set_title('Close Price')
axs[0,0].plot(df1.index,df1['Close'],label="close")
axs[0,0].set_xlabel('Date')
axs[0,0].set_ylabel('Price')
axs[0,0].grid(True)


axs[0,1].set_title('Volume Candle')
axs[0,1].bar(df2.index,df2['Volume'])
axs[0,1].set_xlabel('Date')
axs[0,1].set_ylabel('Volume')
axs[0,1].grid(True)


axs[1,0].set_title('Close Price')
axs[1,0].fill_between(df2.index,df2['Close'],color='skyblue',alpha=0.5)
axs[1,0].plot(df2.index,df2['Close'],label="close",color='blue')
axs[1,0].set_xlabel('Date')
axs[1,0].set_ylabel('Price')
axs[1,0].grid(True)


axs[1,1].set_title("Scatter Plot")
axs[1,1].scatter(df2.index,df2['Open'],color='blue',label="Open")
axs[1,1].scatter(df2.index,df2['Close'],color='orange',label="Close")
axs[1,1].scatter(df2.index,df2['High'],color='green',label="High")
axs[1,1].scatter(df2.index,df2['Low'],color='red',label="Low")
axs[1,1].set_xlabel('Date')
axs[1,1].set_ylabel('Price')
axs[1,1].grid(True)

plt.tight_layout()
plt.show()
```
