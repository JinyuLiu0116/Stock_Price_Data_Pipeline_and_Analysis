```python
import datetime
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import mplfinance as mpl
```

```python
df=yf.download('CRWD')
df.shape
```

```python
df.count()
```

```python
df.tail(10)
```

```python
df=df.round({'Open':4,'Close':4,'High':4,'Low':4,'Adj Close':4})
df.tail(10)
```

```python
plt.figure(figsize=(12,6))
plt.title('Price of CRWD Over The Years', fontsize=18)
plt.plot(df['Close'], label='Close price')

plt.xlabel('Date', fontsize=13)
plt.ylabel('Price In USD ($)', fontsize=13)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title('Price of CRWD In Year 2024', fontsize=18)
plt.plot(df['High'], label='High price', color='green', marker='.')
plt.plot(df['Low'], label='Low price',color='red', marker='.')
plt.xlabel('Date', fontsize=13)
plt.ylabel('Price In USD ($)', fontsize=13)
plt.legend()
plt.show()
```

```python
df0=df['2024-01-01':'2024-07-19']
fig, axs= plt.subplots(2,1, gridspec_kw={"height_ratios": [3,1]}, figsize=(13,7))

mpl.plot(df0, type="candle", ax=axs[0], style="yahoo")
axs[1].plot(df0['Volume'], color='b')
plt.show()
```

```python
df1=df['2024-07-01':'2024-07-18']
fig, axs= plt.subplots(2,1, gridspec_kw={"height_ratios": [3,1]}, figsize=(13,7))

mpl.plot(df1, type="candle", ax=axs[0], style="yahoo")
axs[1].plot(df1['Volume'], color='b')
plt.show()
```
