```python
import datetime
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=yf.download('CRWD', start='2024-01-01',end='2024-07-19')
df
```

```python
df.shape
```

```python
plt.figure(figsize=(12,6))
plt.title('Closing Price History For CRWD', fontsize=18)
plt.plot(df['Close'], label='close price', marker='.')

plt.xlabel('Date', fontsize=13)
plt.ylabel('Price In USD ($)', fontsize=13)
plt.legend()
plt.show()
```

```python
fig, axs= plt.subplots(2,1, gridspec_kw={"height_ratios": [3,1]}, figsize=(13,7))

axs[0].plot(df['Close'],color='blue', marker='.')
axs[1].axhline(y=360, color='r', linestyle='--')
axs[1].axhline(y=300, color='g', linestyle='--')
axs[1].plot(df['Close'])

plt.xlabel('Date', fontsize=13)
plt.ylabel('Price In USD ($)', fontsize=13)
plt.show()
```
