```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=yf.download('CRWD')
df1=df.tail(15)
df1
plt.plot(df)
```

```python
plt.figure(figsize=(10,6))
plt.title('Close Price')
plt.plot(df['Close'],marker='.', markersize=3,label="close")
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.title('Volume Candle')
plt.bar(df1.index,df1['Volume'])
plt.xlabel('Date')
plt.ylabel('Volume')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.title('Close Price')
plt.fill_between(df1.index,df1['Close'],color='skyblue',alpha=0.5)
plt.plot(df1.index,df1['Close'],label="close",color='blue')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.title("Scatter Plot")
plt.scatter(df1.index,df1['Open'],color='blue',label="Open")
plt.scatter(df1.index,df1['Close'],color='orange',label="Close")
plt.scatter(df1.index,df1['High'],color='green',label="High")
plt.scatter(df1.index,df1['Low'],color='red',label="Low")
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
fig, axs=plt.subplots(2,2, figsize=(13,7))

axs[0,0].set_title('Close Price')
axs[0,0].plot(df.index,df['Close'],label="close")
axs[0,0].set_xlabel('Date')
axs[0,0].set_ylabel('Price')
axs[0,0].grid(True)


axs[0,1].set_title('Volume Candle')
axs[0,1].bar(df1.index,df1['Volume'])
axs[0,1].set_xlabel('Date')
axs[0,1].set_ylabel('Volume')
axs[0,1].grid(True)


axs[1,0].set_title('Close Price')
axs[1,0].fill_between(df1.index,df1['Close'],color='skyblue',alpha=0.5)
axs[1,0].plot(df1.index,df1['Close'],label="close",color='blue')
axs[1,0].set_xlabel('Date')
axs[1,0].set_ylabel('Price')
axs[1,0].grid(True)


axs[1,1].set_title("Scatter Plot")
axs[1,1].scatter(df1.index,df1['Open'],color='blue',label="Open")
axs[1,1].scatter(df1.index,df1['Close'],color='orange',label="Close")
axs[1,1].scatter(df1.index,df1['High'],color='green',label="High")
axs[1,1].scatter(df1.index,df1['Low'],color='red',label="Low")
axs[1,1].set_xlabel('Date')
axs[1,1].set_ylabel('Price')
axs[1,1].grid(True)

plt.tight_layout()
plt.show()
```
