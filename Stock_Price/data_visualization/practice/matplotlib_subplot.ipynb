```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x= np.arange(0,100,0.2)
y1=np.sin(x)
y2=x ** 2 +2 * x
y3=np.log(x)
y4=np.tan(x)
```

```python
ax1=plt.subplot(221)
ax2=plt.subplot(222)
ax3=plt.subplot(223)
ax4=plt.subplot(224)

ax1.plot(x,y1,'--')
ax2.plot(x,y2,'g')
ax3.plot(x,y3,'r')
ax4.plot(x,y4,'purple')
plt.tight_layout()
```
