```python
import datetime
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=yf.download('CRWD')
df
```

```python
df=df.round({'Open':4,'Close':4,'High':4,'Low':4,'Adj Close':4})
df
```

```python
df1=selected_columns=df[['Open','High','Low','Close','Adj Close']]
df2=df['Volume']
df3=df['Close']
```

```python
plt.figure(figsize=(12,6))
plt.title("Price of CRWD Over The Years")
plt.plot(df3['2019-06-14':'2024-07-19'],label='Close Price')
plt.ylabel("Price In USD")
plt.xlabel("Time")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.title("Price of CRWD In Year 2024")
plt.plot(df3['2024-01-01':'2024-07-19'],label='Close Price')
plt.ylabel("Price In USD")
plt.xlabel("Time")
plt.legend()
plt.show()
```
