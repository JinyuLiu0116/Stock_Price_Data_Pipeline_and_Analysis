```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-100,100,1)
y1 = 0.5 * x ** 2 + 2 * x
y2 = np.sin(x) * 1000
y3 = np.log(x) * 1000
plt.figure(figsize=(13,7))
plt.plot(x,y1, label='squer function')
plt.plot(x,y2, label='sin function')
plt.plot(x,y3, label='log function')
plt.title("Function Diagram",size=15)
plt.ylabel("Y-axis")
plt.xlabel("X-axis")
plt.legend()
plt.show()
```
